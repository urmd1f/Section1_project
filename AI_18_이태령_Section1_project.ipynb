```python
#원본 사이트에 가면 출고량 단위가 million 인걸 알 수 있음
#M제거, K는 단위 M으로 통일, 실수형으로 전환
import re

def has_MK(inputString) :
  return bool(re.search('[MK]', inputString))

MK_NA_Sales = df[df.NA_Sales.apply(has_MK)].NA_Sales
MK_NA_Sales
```

```python
MK_NA_Sales = MK_NA_Sales.str.replace('[M]', '')
MK_NA_Sales
```

```python
def has_K(inputString) :
  return bool(re.search('[K]', inputString))

K_NA_Sales = MK_NA_Sales[MK_NA_Sales.apply(has_K)]
K_NA_Sales
```

```python
K_NA_Sales = K_NA_Sales.str.replace('[K]','')
K_NA_Sales = K_NA_Sales.astype(float)
K_NA_Sales = K_NA_Sales / 1000
K_NA_Sales
```

```python
for i in range(len(K_NA_Sales)):
  MK_NA_Sales.loc[K_NA_Sales.index[i]] = K_NA_Sales.iloc[i]

MK_NA_Sales
```

```python
for i in range(len(MK_NA_Sales)):
  df.NA_Sales.loc[MK_NA_Sales.index[i]] = MK_NA_Sales.iloc[i]

df.NA_Sales = df.NA_Sales.astype(float)
```
